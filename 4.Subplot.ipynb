```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.hist(y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.plot(x,y,"b:")
plt.subplot(1,2,2)
plt.pie(y)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,"r--")
plt.plot(x,y,"b:")
plt.subplot(2,1,2)
plt.pie(y)
```
