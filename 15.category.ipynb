```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips,x="day",dodge=True,y="total_bill",hue="sex")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x="day",dodge=True,y="total_bill",hue="sex")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x="day",dodge=True,y="total_bill",hue="sex")
sns.swarmplot(data=tips,x="day",dodge=True,y="total_bill",hue="sex",color="0")

plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,kind="swarm")
plt.show()
```
