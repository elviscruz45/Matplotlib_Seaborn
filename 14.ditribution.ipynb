```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.histplot(data=tips,x="tip")
plt.show()
```

```python
sns.histplot(data=tips,x="tip",bins=50)
plt.show()
```

```python
sns.histplot(data=tips,x="tip",bins=50,cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips,x="tip",bins=20,cumulative=False,hue="sex",stat="count",multiple="dodge")
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",cumulative=False,shade=True)
plt.show()
```
