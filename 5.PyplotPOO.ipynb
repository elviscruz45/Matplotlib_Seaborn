```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,"b")
fig.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes2=fig.add_axes([0.2,0.8,0.2,0.2])

axes.plot(x,y,"b")
axes2.plot(y,x,"r")
axes2.set_facecolor("gray")
fig.show()
```
